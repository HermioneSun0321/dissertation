```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import os
import wikipedia

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import Amplifi's knowledge graph data
df_kg = pd.read_csv('kg.csv')

df_text = df_kg[['ent_name','corpus_url']]
df_text
```

```python
df_inv = df_text.iloc[0:6256]
df_wiki = df_text.iloc[6256:8193]
```

```python
list_sent_inv = []

# For investopedia links
for i in range(0,6256):
    URL = df_text['corpus_url'][i]
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    soup_body = str(soup.body)
    
    info = soup.findAll('p', attrs= {'class': 'comp mntl-sc-block finance-sc-block-html mntl-sc-block-html', 'id': 'mntl-sc-block_1-0-1'})
    for data in info: 
        para = data.get_text().strip()
        sent = para.partition('.')[0] + '.'
    list_sent_inv.append(sent)

list_sent_inv
```

```python
df_inv['sent'] = list_sent_inv
df_inv
```

```python
list_head = []
list_sent_wiki = []

# For Wikipedia links
for i in range(6256,8193):
    URL = df_text['corpus_url'][i]
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    soup_body = str(soup.body)


    headings = soup.findAll('h1', attrs= {'class': 'firstHeading mw-first-heading', 'id': 'firstHeading'})
    for data in headings: 
        head = data.get_text().strip()
    list_head.append(head)

    for i in range(0,len(list_head)):
        try:
            sent = wikipedia.summary(list_head[i], sentences=1)
        except (wikipedia.DisambiguationError, wikipedia.PageError):
            sent = 'Error'
    list_sent_wiki.append(sent)

list_sent_wiki
```

```python
df_wiki['sent'] = list_sent_wiki
df_wiki
```

```python
list_sent = list_sent_inv + list_sent_wiki
df_text['sent'] = list_sent
df_text['sent'] = df_text['sent'].astype('str')
df_text
```
